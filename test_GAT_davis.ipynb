## train and test GAT model on the URV database with n_folds = 10.

```python
# test of GAT model on URV dataset with n_folds = 10
from URV_Test import Testfold
import models.gat as GAT

# Create an instance of the model
model = GAT.GATNet()
dataset = 'urv'
cuda_name = "cuda:0"

test_mse_list = Testfold(model, dataset, cuda_name, TEST_BATCH_SIZE = 1, n_folds = 10)
```
