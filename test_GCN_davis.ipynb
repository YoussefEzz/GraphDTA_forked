## train and test GCN model with davis database.

```python
# train of GCN model with davis dataset using 1000 epochs
import sys
import os

from URV_Train import Train

import models.gcn as GCN

model = Train(dataset = 'davis', model_type = GCN.GCNNet(), cuda_name = "cuda:0", TRAIN_BATCH_SIZE = 512, TEST_BATCH_SIZE = 512, LR = 0.0005, NUM_EPOCHS = 1000)
```

```python
# test of GCN model with davis dataset
from URV_Test import Test
import torch
import models.gcn as GCN

# Create an instance of the model
model = GCN.GCNNet()

# Load the state dictionary from the file
model_file_name = 'model_GCNNet_davis.model'
state_dict = torch.load(model_file_name)

# Load the state dictionary into the model
model.load_state_dict(state_dict)

# Optionally, set the model to evaluation mode
model.eval()
test_mse = Test(model, dataset = 'davis', cuda_name = "cuda:0", TEST_BATCH_SIZE = 512 )
```
