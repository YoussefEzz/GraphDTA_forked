```python
# train of GIN CONV model with davis dataset
from URV_Train import Train
import models.ginconv as GINCONV

model = Train(dataset = 'davis', model_type = GINCONV.GINConvNet(), cuda_name = "cuda:0", TRAIN_BATCH_SIZE = 512, TEST_BATCH_SIZE = 512, LR = 0.0005, NUM_EPOCHS = 1000)
```

```python
# test of GINCONV model on davis dataset
from URV_Test import Test
import torch
import models.ginconv as GINCONV

# Create an instance of the model
model = GINCONV.GINConvNet()

# Load the state dictionary from the file
model_file_name = 'model_GINConvNet_davis.model'
state_dict = torch.load(model_file_name)

# Load the state dictionary into the model
model.load_state_dict(state_dict)

# Optionally, set the model to evaluation mode
model.eval()
test_mse = Test(model, dataset = 'davis', cuda_name = "cuda:0", TEST_BATCH_SIZE = 512 )

```
