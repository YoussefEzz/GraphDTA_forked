```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

davis_datapath = "data"
davis_train_filepath = davis_datapath + '/davis_train.csv'
davis_test_filepath = davis_datapath + '/davis_test.csv'

# Read the train Affinity TXT file into a dataframe
train_dataframe = pd.read_csv(davis_train_filepath,sep=r',')  # Use the appropriate delimiter if your file is not tab-separated

# Read the train TXT file into a dataframe
test_dataframe = pd.read_csv(davis_test_filepath,sep=r',')  # Use the appropriate delimiter if your file is not tab-separated

# column_names = train_dataframe.columns.tolist()
# print(column_names)


# Select the train affinity column
train_affinity = train_dataframe.iloc[:, 2]
print(len(train_affinity))

# Select the test affinity column
test_affinity = test_dataframe.iloc[:, 2]
print(len(test_affinity))

# Concatenate the train and test affinity columns
affinity = pd.concat([train_affinity, test_affinity], ignore_index=True)
print(len(affinity))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(10)
data = affinity
print(data)

# Create a boxplot
sns.boxplot(data=data, width = 0.07, fill= False)

# Add the points
# For visual clarity, jitter the points along the x-axis
x = np.random.normal(0, 0.01, size=len(data)) # add some jitter
plt.scatter(x, data, color='red', alpha=0.6)

# Set labels and title
plt.xlabel(f'(size = {len(affinity)}), jitter = {0.01}')
plt.ylabel('affinity')
plt.title('Box Plot for Drug Protein affinity with jitter = 0.01')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of Kd values
Kd_range = np.linspace(1, 10000)

# Calculate the pkd values
pkd = -np.log10(Kd_range / 1e9)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(Kd_range, pkd)

# Set the x-axis scale to logarithmic
#ax.set_xscale('log')

# Set the axis labels and title
ax.set_xlabel('Kd in nanoMolar')
ax.set_ylabel('pkd')
ax.set_title('Plot of pkd = -log_10 (Kd/1e9)')

# Show the plot
plt.show()
```
