```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

kiba_datapath = "data"
kiba_train_filepath = kiba_datapath + '/kiba_train.csv'
kiba_test_filepath = kiba_datapath + '/kiba_test.csv'

# Read the train Affinity TXT file into a dataframe
train_dataframe = pd.read_csv(kiba_train_filepath,sep=r',')  # Use the appropriate delimiter if your file is not tab-separated

# Read the train TXT file into a dataframe
test_dataframe = pd.read_csv(kiba_test_filepath,sep=r',')  # Use the appropriate delimiter if your file is not tab-separated

# column_names = train_dataframe.columns.tolist()
# print(column_names)

# Select the train smiles column
train_smiles = train_dataframe.iloc[:, 0]
print(len(train_smiles))

# Select the test smiles column
test_smiles = test_dataframe.iloc[:, 0]
print(len(test_smiles))

# Concatenate the train and test smiles columns
smiles = pd.concat([train_smiles, test_smiles], ignore_index=True)
smiles_lengths = [len(smile) for smile in smiles]
print(len(smiles))


# Select the train protein sequence column
train_protein = train_dataframe.iloc[:, 1]
print(len(train_protein))

# Select the test smiles column
test_protein = test_dataframe.iloc[:, 1]
print(len(test_protein))

# Concatenate the train and test smiles columns
proteins = pd.concat([train_protein, test_protein], ignore_index=True)
proteins_lengths = [len(protein) for protein in proteins]
print(len(proteins))


# Select the train affinity column
train_affinity = train_dataframe.iloc[:, 2]
print(len(train_affinity))

# Select the test affinity column
test_affinity = test_dataframe.iloc[:, 2]
print(len(test_affinity))

# Concatenate the train and test affinity columns
affinity = pd.concat([train_affinity, test_affinity], ignore_index=True)
print(len(affinity))


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(10)
data = affinity
print(data)

# Create a boxplot
sns.boxplot(data=data, width = 0.07, fill= False)

# Add the points
# For visual clarity, jitter the points along the x-axis
x = np.random.normal(0, 0.01, size=len(data)) # add some jitter
plt.scatter(x, data, color='red', alpha=0.6)

# Set labels and title
plt.xlabel(f'(size = {len(affinity)}), jitter = {0.01}')
plt.ylabel('affinity(KIBA score)')
plt.title('Box Plot for Drug Protein affinity kiba score with jitter = 0.01')

# Show plot
plt.show()
```
